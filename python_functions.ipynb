```python
def hitung_total(*args):
    total = sum(args)
    return total

print(hitung_total(5, 6, 8, 10))
```

```python
def data_diri(**kwargs):
  info = ""
  for key, value in kwargs.items():
    print(value)
    info += key + ': ' + str(value) + ", "
  return info

print(data_diri(nama="Test", usia=37, pekerjaan="Python Programmer"))
```
